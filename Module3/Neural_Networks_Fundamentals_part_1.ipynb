# 1. Biological Inspiration


The human brain contains billions of neurons connected through synapses.
Learning occurs by strengthening or weakening these connections.

Biological Components:

- Dendrites → Receive signals
- Cell Body (Soma) → Processing
- Axon → Output transmission
- Synapse → Connection mechanism

Neural networks attempt to replicate this learning mechanism computationally.


# 2. Artificial Neurons


An artificial neuron receives inputs, multiplies them by weights, adds bias, and applies an activation function.

Mathematical Model:

$$
z=\sum_{i=1}^{n} w_i x_i + b
$$

$$
y=f(z)
$$

where:

- $x_i$ = inputs
- $w_i$ = weights
- $b$ = bias
- $f$ = activation function


# 3. Perceptron


The Perceptron is the simplest neural network model introduced by Frank Rosenblatt.

Decision Rule:

$$
y =
\begin{cases}
1 & z \geq 0 \\
0 & z < 0
\end{cases}
$$

It can solve linearly separable problems.


# 4. Weights and Biases


Weights determine feature importance.

Bias shifts the decision boundary.

Neuron Equation:

$$
z = W^TX + b
$$

Large positive weights increase influence while negative weights suppress influence.


# 5. Activation Functions


Activation functions introduce non-linearity.

### Sigmoid

$$
\sigma(x)=\frac{1}{1+e^{-x}}
$$

### Tanh

$$
\tanh(x)=\frac{e^x-e^{-x}}{e^x+e^{-x}}
$$

### ReLU

$$
ReLU(x)=max(0,x)
$$

### Softmax

$$
P(y_i)=\frac{e^{z_i}}{\sum_j e^{z_j}}
$$


# 6. Forward Propagation


Forward propagation computes outputs layer-by-layer.

Hidden Layer:

$$
z^{(1)}=W^{(1)}X+b^{(1)}
$$

$$
a^{(1)}=f(z^{(1)})
$$

Output Layer:

$$
z^{(2)}=W^{(2)}a^{(1)}+b^{(2)}
$$

$$
\hat y=f(z^{(2)})
$$


# 7. Loss Functions


Loss functions measure prediction error.

### Mean Squared Error

$$
MSE=\frac{1}{n}\sum(y-\hat y)^2
$$

### Binary Cross Entropy

$$
L=-(y\log(\hat y)+(1-y)\log(1-\hat y))
$$

### Categorical Cross Entropy

$$
L=-\sum y_i \log(\hat y_i)
$$


# 8. Binary Classification


Binary classification predicts one of two classes.

Examples:

- Spam Detection
- Fraud Detection
- Disease Prediction

Output layer usually uses Sigmoid activation.


# 9. Multiclass Classification


Predicts one class among multiple categories.

Examples:

- Handwritten Digit Recognition
- Object Recognition
- Language Identification

Output layer uses Softmax activation.


# 10. Regression Neural Networks


Regression predicts continuous numerical values.

Examples:

- House Price Prediction
- Sales Forecasting
- Energy Consumption Prediction

Output layer often uses Linear Activation.


# NumPy Implementation: Artificial Neuron

```

import numpy as np

X = np.array([2,3,4])
W = np.array([0.5,0.2,0.8])
b = 1

output = np.dot(X,W)+b

print("Neuron Output:", output)
```

# Activation Function Visualizations

```

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,200)

sigmoid = 1/(1+np.exp(-x))
relu = np.maximum(0,x)

plt.figure(figsize=(8,4))
plt.plot(x,sigmoid,label='Sigmoid')
plt.plot(x,relu,label='ReLU')
plt.legend()
plt.title('Activation Functions')
plt.grid(True)
plt.show()
```

# Binary Classification Example

```

import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.array([1,2])
w = np.array([0.6,0.8])
b = -1

prediction = sigmoid(np.dot(x,w)+b)

print("Probability:", prediction)
```

# Multiclass Classification Example

```

import numpy as np

scores = np.array([2.1, 1.5, 0.3])

exp_scores = np.exp(scores)
softmax = exp_scores/np.sum(exp_scores)

print("Class Probabilities:", softmax)
```
