# 1. Forward Propagation

## What Is Forward Propagation?

Forward propagation is the process of moving information from inputs to outputs.

### Single Neuron

\[
z = \sum_i w_i x_i + b
\]

\[
a=f(z)
\]

### Matrix Representation

\[
Z=XW+b
\]

\[
A=f(Z)
\]

### Multi-Layer Network

\[
Z_1=XW_1+b_1
\]

\[
A_1=f(Z_1)
\]

\[
Z_2=A_1W_2+b_2
\]

\[
A_2=f(Z_2)
\]

Each layer transforms representations learned from the previous layer.


```

import numpy as np

X=np.array([1,2,3])
W=np.array([0.4,0.3,0.8])
b=0.5

z=np.dot(X,W)+b
print("Forward Output:",z)
```


## Numerical Example

Inputs = [1,2,3]

Weights = [0.4,0.3,0.8]

Bias = 0.5

Output:

(1×0.4)+(2×0.3)+(3×0.8)+0.5 = 3.9


```

import numpy as np

def relu(x):
    return np.maximum(0,x)

X=np.array([[1,2]])

W1=np.array([[0.5,-0.3],
             [0.8,0.2]])

b1=np.array([0.1,0.1])

Z1=X.dot(W1)+b1
A1=relu(Z1)

print("Hidden Layer:",A1)
```


# 2. Loss Functions

## Why Loss Functions Exist

Predictions are rarely perfect.

Loss functions measure the difference between:

Actual Output \(y\)

Predicted Output \(\hat y\)

## Mean Squared Error

\[
MSE=\frac1n\sum(y-\hat y)^2
\]

Penalizes large errors heavily.

## Mean Absolute Error

\[
MAE=\frac1n\sum|y-\hat y|
\]

More robust to outliers.

## Binary Cross Entropy

\[
BCE=-[y\log(p)+(1-y)\log(1-p)]
\]

Used for binary classification.

## Categorical Cross Entropy

\[
CCE=-\sum y_i \log(p_i)
\]

Used for multiclass classification.


```

import numpy as np
import matplotlib.pyplot as plt

errors=np.linspace(-5,5,200)

mse=errors**2
mae=np.abs(errors)

plt.figure(figsize=(6,4))
plt.plot(errors,mse,label="MSE")
plt.plot(errors,mae,label="MAE")
plt.legend()
plt.grid(True)
plt.title("Loss Function Comparison")
plt.show()
```


# Loss Function Comparison

| Task | Preferred Loss |
|--------|---------------|
| Regression | MSE / MAE |
| Binary Classification | BCE |
| Multiclass Classification | Categorical Cross Entropy |



# 3. Binary Classification

Binary classification predicts one of two classes.

Examples:

- Spam vs Not Spam
- Fraud vs Legitimate
- Disease vs Healthy

Architecture:

Input → Hidden Layer → Sigmoid Output

Probability Interpretation:

- 0.93 → strong confidence
- 0.50 → uncertain
- 0.12 → likely negative class


```

import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

x=np.array([2,1])
w=np.array([0.8,0.4])
b=-0.5

p=sigmoid(np.dot(x,w)+b)
print("Probability:",p)
```


## Decision Threshold

Common threshold:

\[
p > 0.5 \rightarrow Class\;1
\]

\[
p \le 0.5 \rightarrow Class\;0
\]

Thresholds may be adjusted for precision/recall trade-offs.



# 4. Multiclass Classification

Examples:

- Digit Recognition
- Animal Classification
- Disease Categories

Uses Softmax outputs and Cross Entropy Loss.

## One-Hot Encoding

Cat = [1,0,0]

Dog = [0,1,0]

Bird = [0,0,1]


```

import numpy as np

scores=np.array([2.5,1.2,0.4])

exp=np.exp(scores)
softmax=exp/np.sum(exp)

print("Probabilities:",softmax)
print("Sum:",np.sum(softmax))
```


# 5. Regression Neural Networks

Regression predicts continuous values.

Examples:

- House Prices
- Temperature Forecasts
- Sales Forecasts

Regression networks typically use a linear output layer rather than Softmax.


```

import numpy as np

x=np.array([1200,3])
w=np.array([150,10000])
b=50000

price=np.dot(x,w)+b

print("Predicted Price:",price)
```

# 11. Summary

## Key Takeaways

- Forward propagation generates predictions.
- Loss functions quantify errors.
- BCE and CCE are classification losses.
- MSE and MAE are regression losses.
- Softmax produces probability distributions.
- Regression predicts continuous values.

## Formula Cheat Sheet

MSE = mean((y-yhat)^2)

MAE = mean(|y-yhat|)

BCE = -(ylog(p)+(1-y)log(1-p))

Softmax = exp(z)/sum(exp(z))
