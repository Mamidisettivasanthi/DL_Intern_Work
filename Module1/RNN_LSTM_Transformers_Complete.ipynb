## Sequential Data

Text, speech, time-series, DNA sequences and sensor streams contain temporal dependencies.

## Why Feedforward Networks Fail on Sequences

Feedforward networks have no memory and cannot naturally model temporal context.

## Recurrent Neural Networks (RNN)

RNNs maintain hidden states that carry information across time steps.

## Vanishing Gradient Problem

Gradients shrink across long sequences, making long-term dependencies difficult to learn.

## LSTM Architecture

LSTMs use forget, input, and output gates to preserve long-term information.

## GRU Architecture

GRUs simplify LSTMs using update and reset gates.

## Attention Mechanism

Attention lets models focus on relevant tokens dynamically.

## Self-Attention

Tokens attend to other tokens within the same sequence.

## Multi-Head Attention

Multiple attention heads learn different relationships in parallel.

## Transformer Architecture

Transformers use self-attention and feedforward layers instead of recurrence.

## Level 2: LSTM Example (PyTorch)

```python
import torch
import torch.nn as nn

class TextClassifier(nn.Module):

    def __init__(self):
        super().__init__()

        self.embedding = nn.Embedding(
            num_embeddings=5000,
            embedding_dim=64
        )

        self.lstm = nn.LSTM(
            input_size=64,
            hidden_size=64,
            batch_first=True
        )

        self.fc = nn.Linear(
            64,
            1
        )

    def forward(self, x):

        x = self.embedding(x)

        _, (hidden, _) = self.lstm(x)

        output = torch.sigmoid(
            self.fc(hidden[-1])
        )

        return output


model = TextClassifier()

print(model)
```

## Level 4: Attention Mechanism

```python
def softmax(x):
    e=np.exp(x-np.max(x))
    return e/e.sum()

Q=np.random.rand(4,4)
K=np.random.rand(4,4)
V=np.random.rand(4,4)

scores=np.dot(Q,K.T)
weights=softmax(scores)
output=np.dot(weights,V)

print(output.shape)
```

## Level 5: Mini Transformer

```python
import torch
import torch.nn as nn

class TransformerBlock(nn.Module):

    def __init__(
        self,
        embed_dim=128,
        num_heads=8,
        ff_hidden=512,
        dropout=0.1
    ):
        super().__init__()

        self.attention = nn.MultiheadAttention(
            embed_dim=embed_dim,
            num_heads=num_heads,
            dropout=dropout,
            batch_first=True
        )

        self.norm1 = nn.LayerNorm(embed_dim)

        self.feed_forward = nn.Sequential(
            nn.Linear(embed_dim, ff_hidden),
            nn.ReLU(),
            nn.Linear(ff_hidden, embed_dim)
        )

        self.norm2 = nn.LayerNorm(embed_dim)

        self.dropout = nn.Dropout(dropout)

    def forward(self, x):

        # Multi-Head Self Attention
        attn_output, _ = self.attention(
            x, x, x
        )

        # Residual Connection + LayerNorm
        x = self.norm1(
            x + self.dropout(attn_output)
        )

        # Feed Forward Network
        ff_output = self.feed_forward(x)

        # Residual Connection + LayerNorm
        x = self.norm2(
            x + self.dropout(ff_output)
        )

        return x
```

## Mathematics of Attention

Attention(Q,K,V)

= softmax(QKᵀ / √dₖ)V

Where:

- Q = Queries
- K = Keys
- V = Values
- dₖ = Key dimension


## Vanishing Gradient Problem

Repeated multiplication of gradients:

∂L/∂W

becomes extremely small across many time steps.

Solutions:
- LSTM
- GRU
- Residual Connections
- Transformers
