## Why Training is Needed
Training enables a neural network to learn patterns by minimizing prediction error.

## Training Pipeline
Data Collection → Preprocessing → Forward Pass → Loss Computation → Backpropagation → Weight Update → Evaluation

## Loss Functions

### Mean Squared Error (MSE)
$MSE=\frac{1}{n}\sum(y-\hat y)^2$

### Cross Entropy
$L=-\sum y\log(\hat y)$


## Gradient Descent

Weight Update Rule:

$w_{new}=w_{old}-\eta \frac{\partial L}{\partial w}$

where:
- η = learning rate
- L = loss


## Backpropagation Intuition

Backpropagation computes how much each weight contributed to the final error and adjusts it accordingly.

### Chain Rule
If:

$z=f(y)$ and $y=g(x)$

then

$\frac{dz}{dx}=\frac{dz}{dy}\frac{dy}{dx}$

## Level 1: Gradient Descent From Scratch

```python
w=10
lr=0.1
history=[]

for _ in range(50):
    grad=2*w
    w=w-lr*grad
    history.append(w)

plt.plot(history)
plt.title("Gradient Descent Convergence")
plt.show()
print("Final Weight:",w)
```

## Level 2: Linear Regression using Gradient Descent

```python
X=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])

w=0.0
b=0.0
lr=0.01

for epoch in range(1000):
    pred=w*X+b
    dw=(-2/len(X))*np.sum(X*(y-pred))
    db=(-2/len(X))*np.sum(y-pred)

    w-=lr*dw
    b-=lr*db

print("Weight:",w)
print("Bias:",b)
```

## Level 3: Backpropagation using NumPy

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

X=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([[0],[1],[1],[0]])

np.random.seed(42)
W1=np.random.randn(2,4)
W2=np.random.randn(4,1)

for epoch in range(2000):

    h=sigmoid(np.dot(X,W1))
    out=sigmoid(np.dot(h,W2))

    error=y-out

    d_out=error*out*(1-out)

    d_hidden=np.dot(d_out,W2.T)*h*(1-h)

    W2+=0.1*np.dot(h.T,d_out)
    W1+=0.1*np.dot(X.T,d_hidden)

print(out)
```

## Training vs Validation Loss

```python
epochs=np.arange(1,51)
train=1/(epochs**0.5)
val=train+0.002*epochs

plt.plot(epochs,train,label='Train')
plt.plot(epochs,val,label='Validation')
plt.legend()
plt.title("Training vs Validation Loss")
plt.show()
```

## Level 4: TensorFlow Training Example

```python
import tensorflow as tf
model=tf.keras.Sequential([
     tf.keras.layers.Dense(8,activation='relu',input_shape=(2,)),
     tf.keras.layers.Dense(1,activation='sigmoid')
])
model.compile(optimizer='adam',
               loss='binary_crossentropy',
               metrics=['accuracy'])
model.summary()
```
