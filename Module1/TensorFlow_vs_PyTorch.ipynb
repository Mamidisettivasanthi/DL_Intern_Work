## History

### TensorFlow
- Released by Google (2015)
- Production-focused ecosystem
- Strong deployment support

### PyTorch
- Released by Facebook/Meta (2016)
- Research-friendly design
- Pythonic and flexible


## Dynamic vs Static Graphs

### TensorFlow
Originally static graph, now eager execution by default.

### PyTorch
Dynamic computation graph by design.


## Ecosystem Comparison

| Feature | TensorFlow | PyTorch |
|----------|------------|---------|
| Research | Good | Excellent |
| Production | Excellent | Very Good |
| Mobile | Strong | Improving |
| Deployment | TensorFlow Serving | TorchServe |


````python
import numpy as np
import matplotlib.pyplot as plt
````

## TensorFlow Neural Network

````python
import tensorflow as tf

tf_model = tf.keras.Sequential([
    tf.keras.layers.Dense(
        16,
        activation='relu',
        input_shape=(4,)
    ),
    tf.keras.layers.Dense(
        8,
        activation='relu'
    ),
    tf.keras.layers.Dense(
        1,
        activation='sigmoid'
    )
])

tf_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

tf_model.summary()
````

## PyTorch Neural Network

````python
import torch
import torch.nn as nn

class Net(nn.Module):

    def __init__(self):
        super().__init__()

        self.net = nn.Sequential(
            nn.Linear(4, 16),
            nn.ReLU(),
            nn.Linear(16, 8),
            nn.ReLU(),
            nn.Linear(8, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.net(x)

model = Net()

print(model)
````

## Syntax Comparison

TensorFlow:
```python
model.fit(X,y)
```

PyTorch:
```python
loss.backward()
optimizer.step()
```


## Training Comparison

TensorFlow:
- High-level API

PyTorch:
- More control
- Explicit training loop


## Evaluation Comparison

TensorFlow:
```python

```

PyTorch:


## Training Curves Comparison

````python
epochs=np.arange(1,51)

tf_loss=np.exp(-0.08*epochs)
torch_loss=np.exp(-0.075*epochs)

plt.plot(epochs,tf_loss,label='TensorFlow')
plt.plot(epochs,torch_loss,label='PyTorch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Curves Comparison')
plt.legend()
plt.show()
````
