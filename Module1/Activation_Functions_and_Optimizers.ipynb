## Step Function

### Intuition
Step Function activation.

### Formula
f(x)=1 if x>=0 else 0

### Derivative
Not differentiable

### Advantages
- Simple

### Limitations
- No gradient learning

### Use Cases
- Early perceptrons

## Sigmoid

### Intuition
Sigmoid activation.

### Formula
1/(1+e^-x)

### Derivative
σ(x)(1-σ(x))

### Advantages
- Probabilistic output

### Limitations
- Vanishing gradients

### Use Cases
- Binary classification

## Tanh

### Intuition
Tanh activation.

### Formula
tanh(x)

### Derivative
1-tanh²(x)

### Advantages
- Zero centered

### Limitations
- Still vanishes

### Use Cases
- RNNs

## ReLU

### Intuition
ReLU activation.

### Formula
max(0,x)

### Derivative
1 or 0

### Advantages
- Fast

### Limitations
- Dying ReLU

### Use Cases
- Hidden layers

## Leaky ReLU

### Intuition
Leaky ReLU activation.

### Formula
max(ax,x)

### Derivative
1 or a

### Advantages
- Avoids dead neurons

### Limitations
- Needs alpha

### Use Cases
- CNNs

## ELU

### Intuition
ELU activation.

### Formula
ELU(x)

### Derivative
Piecewise

### Advantages
- Smooth negative region

### Limitations
- More compute

### Use Cases
- Deep nets

## GELU

### Intuition
GELU activation.

### Formula
Gaussian Error Linear Unit

### Derivative
Approx derivative

### Advantages
- State of art

### Limitations
- Expensive

### Use Cases
- Transformers

## Softmax

### Intuition
Softmax activation.

### Formula
exp(xi)/Σexp(x)

### Derivative
Jacobian

### Advantages
- Probabilities

### Limitations
- Output layer only

### Use Cases
- Multiclass output

```
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-5,5,500)

def step(x): return np.where(x>=0,1,0)
def sigmoid(x): return 1/(1+np.exp(-x))
def tanh(x): return np.tanh(x)
def relu(x): return np.maximum(0,x)
def leaky_relu(x,a=0.01): return np.where(x>0,x,a*x)
def elu(x,a=1): return np.where(x>0,x,a*(np.exp(x)-1))
def gelu(x):
    return 0.5*x*(1+np.tanh(np.sqrt(2/np.pi)*(x+0.044715*x**3)))
```

## Activation Curves

```
funcs=[('Step',step),('Sigmoid',sigmoid),('Tanh',tanh),('ReLU',relu),('Leaky ReLU',leaky_relu),('ELU',elu),('GELU',gelu)]
for n,f in funcs:
    plt.figure(figsize=(5,3))
    plt.plot(x,f(x))
    plt.title(n)
    plt.grid(True)
    plt.show()
```

## Derivative Curves

```
sig=sigmoid(x)*(1-sigmoid(x))
tan=1-np.tanh(x)**2
relu_d=np.where(x>0,1,0)

plt.plot(x,sig,label='Sigmoid')
plt.plot(x,tan,label='Tanh')
plt.plot(x,relu_d,label='ReLU')
plt.legend()
plt.show()
```

# Optimizers

## SGD
w = w - lr * grad

## Momentum
Uses velocity to smooth updates.

## RMSProp
Adaptive learning rates.

## Adam
Momentum + RMSProp.

## AdamW
Adam with decoupled weight decay.


```
def sgd(w,g,lr=0.01):
    return w-lr*g

def momentum(w,g,v,lr=0.01,b=0.9):
    v=b*v+lr*g
    return w-v,v

def rmsprop(w,g,s,lr=0.01,b=0.9):
    s=b*s+(1-b)*(g*g)
    return w-lr*g/(np.sqrt(s)+1e-8),s

def adam(w,g,m,v,t,lr=0.01,b1=0.9,b2=0.999):
    m=b1*m+(1-b1)*g
    v=b2*v+(1-b2)*(g*g)
    mh=m/(1-b1**t)
    vh=v/(1-b2**t)
    w=w-lr*mh/(np.sqrt(vh)+1e-8)
    return w,m,v
```

## SGD vs Momentum vs Adam

```
epochs=np.arange(1,101)
sgd_loss=np.exp(-0.03*epochs)
mom_loss=np.exp(-0.05*epochs)
adam_loss=np.exp(-0.08*epochs)

plt.plot(epochs,sgd_loss,label='SGD')
plt.plot(epochs,mom_loss,label='Momentum')
plt.plot(epochs,adam_loss,label='Adam')
plt.legend()
plt.title('Convergence Comparison')
plt.show()
```
