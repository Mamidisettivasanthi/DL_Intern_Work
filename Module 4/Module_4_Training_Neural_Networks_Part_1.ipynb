# 4. Understanding Error

Error = Actual Value − Predicted Value

Example:

Actual House Price = ₹50,00,000

Predicted = ₹47,00,000

Error = ₹3,00,000

Reducing error improves model performance.


```

import matplotlib.pyplot as plt

actual=[50]
good=[49]
bad=[30]

plt.figure(figsize=(5,4))
plt.scatter([1],[actual[0]],label='Actual')
plt.scatter([2],[good[0]],label='Good Prediction')
plt.scatter([3],[bad[0]],label='Bad Prediction')
plt.legend()
plt.title("Prediction Error Comparison")
plt.show()
```


# 6. Gradient Descent

## Historical Background

Gradient-based optimization became the foundation of modern machine learning and deep learning.

## Intuition

Imagine standing on a mountain and trying to reach the valley.

The steepest downhill direction is given by the negative gradient.

## Mathematical Derivation

Given loss:

\[
L(w)
\]

Gradient:

\[
\frac{dL}{dw}
\]

Update Rule:

\[
w_{new}=w_{old}-\eta\frac{dL}{dw}
\]

Where:

- \(w\) = parameter
- \(\eta\) = learning rate
- \(dL/dw\) = gradient



## Geometric Interpretation

Gradient points toward steepest increase.

Gradient Descent moves in the opposite direction to reduce loss.


```

import numpy as np

def loss(w):
    return w**2

def gradient(w):
    return 2*w

w=5.0
lr=0.1

history=[]

for _ in range(20):
    history.append(w)
    w=w-lr*gradient(w)

print("Optimized Weight:",w)
```

```

import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.plot(history,'o-')
plt.title("Gradient Descent Convergence")
plt.xlabel("Iteration")
plt.ylabel("Weight Value")
plt.grid(True)
plt.show()
```


## Gradient Descent Algorithm

1. Initialize parameters.
2. Compute loss.
3. Compute gradients.
4. Update parameters.
5. Repeat until convergence.



# Numerical Example

Loss:

L(w)=w²

Initial:

w=4

Gradient:

2w=8

Learning Rate:

η=0.1

Update:

w_new=4−0.1×8=3.2

Repeating updates gradually converges toward zero.



# 7. Learning Rate

Learning rate determines step size during optimization.



## Small Learning Rate

Advantages:
- Stable updates

Disadvantages:
- Slow training



## Large Learning Rate

Advantages:
- Faster progress

Disadvantages:
- May overshoot minima
- Can diverge


```

import numpy as np
import matplotlib.pyplot as plt

iterations=np.arange(20)

small=np.exp(-0.1*iterations)
ideal=np.exp(-0.3*iterations)
large=np.abs(np.cos(iterations))

plt.figure(figsize=(7,4))
plt.plot(small,label='Small LR')
plt.plot(ideal,label='Ideal LR')
plt.plot(large,label='Large LR')
plt.legend()
plt.grid(True)
plt.title("Learning Rate Effects")
plt.show()
```


## Learning Rate Scheduling

### Step Decay

Reduce learning rate after fixed intervals.

### Exponential Decay

Gradually decreases learning rate.

### Cosine Annealing

Uses cosine-shaped schedule.



# 8. Convergence Analysis

## Signs of Convergence

- Stable loss values
- Small gradients

## Signs of Divergence

- Increasing loss
- Oscillation
- Exploding updates

Challenges:

- Overshooting
- Plateaus
- Slow Learning



# 9. Challenges of Gradient Descent

### Local Minima

May trap optimization.

### Saddle Points

Gradient near zero but not optimal.

### Flat Regions

Learning slows dramatically.

### Noisy Gradients

Updates become unstable.

Modern optimizers help overcome these challenges.
