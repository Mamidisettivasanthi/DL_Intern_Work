# 1. Why Weight Initialization Matters

Initialization determines the starting point of optimization.

Poor initialization can:

- Slow learning
- Cause gradient problems
- Prevent convergence

Good initialization promotes stable gradient flow and faster learning.



# 2. Zero Initialization

All weights initialized to zero:

\[
W = 0
\]

Problem:

All neurons produce identical outputs and gradients.

This symmetry prevents meaningful learning.

Therefore zero initialization fails for deep networks.


```

import numpy as np

W=np.zeros((3,3))
print(W)
```


# 3. Random Initialization

Weights initialized randomly.

Benefits:

- Breaks symmetry
- Enables diverse feature learning

Drawbacks:

- Too large → exploding activations
- Too small → vanishing activations


```

import numpy as np

W=np.random.randn(3,3)*0.01
print(W)
```


# 4. Xavier Initialization

Historical Background:

Introduced by Glorot and Bengio.

Variance:

\[
Var(W)=\frac{1}{n}
\]

Designed to preserve activation variance across layers.


```

import numpy as np

n=100
W=np.random.randn(n,n)*np.sqrt(1/n)

print(W.shape)
```


# 5. He Initialization

Developed for ReLU networks.

Variance:

\[
Var(W)=\frac{2}{n}
\]

Better suited for deep ReLU architectures.


```

import numpy as np

n=100
W=np.random.randn(n,n)*np.sqrt(2/n)

print(W.shape)
```


# 6. Weight Initialization Comparison

| Method | Symmetry Breaking | Deep Networks |
|----------|----------|----------|
| Zero | No | Poor |
| Random | Yes | Moderate |
| Xavier | Yes | Good |
| He | Yes | Excellent |



# 7. Why Basic Gradient Descent Is Not Enough

Challenges:

Modern optimizers improve training efficiency.



# 8. Optimizers

## SGD

Update:

\[
W = W - \eta \nabla L
\]

Simple but can be slow.



## Momentum

Velocity:

\[
v_t=\beta v_{t-1}+(1-\beta)\nabla L
\]

Update:

\[
W=W-\eta v_t
\]

Accelerates learning.



## Nesterov Momentum

Uses look-ahead gradients.

Provides more accurate updates and often converges faster.



## AdaGrad

Adaptive learning rate:

\[
G_t=G_{t-1}+g_t^2
\]

Update:

\[
W=W-\frac{\eta}{\sqrt{G_t+\epsilon}}g_t
\]

Works well for sparse data.



## RMSProp

Running average:

\[
E[g^2]_t=\rho E[g^2]_{t-1}+(1-\rho)g_t^2
\]

Provides adaptive step sizes.



## Adam

Combines Momentum and RMSProp.

First Moment:

\[
m_t
\]

Second Moment:

\[
v_t
\]

Update:

\[
W=W-\eta\frac{\hat m_t}{\sqrt{\hat v_t}+\epsilon}
\]


# 9. AdamW

AdamW decouples weight decay from gradient updates.

Benefits:

- Better regularization
- Better generalization

Widely used in Transformers and LLMs.


# 11. Learning Curves

Learning curves track:

- Training Loss
- Validation Loss

They help diagnose model behavior.


```

import numpy as np
import matplotlib.pyplot as plt

epochs=np.arange(1,21)

train=np.exp(-0.2*epochs)
val=np.exp(-0.15*epochs)+0.05

plt.figure(figsize=(6,4))
plt.plot(epochs,train,label='Training')
plt.plot(epochs,val,label='Validation')
plt.legend()
plt.grid(True)
plt.title("Learning Curves")
plt.show()
```


# 12. Underfitting

Symptoms:

- High training loss
- High validation loss

Causes:

# 13. Overfitting

Symptoms:
