# 1. Training Data Processing

Modern datasets contain millions of examples.

Processing all data simultaneously is often impossible because:

- Memory limitations
- Computational constraints
- Training efficiency concerns

Therefore data is divided into manageable batches.



# 2. Batch Learning

## Full Batch Gradient Descent

Uses the entire dataset to compute gradients.

Advantages:
- Stable updates

Disadvantages:
- Slow
- High memory usage



## Stochastic Gradient Descent (SGD)

Uses one training example at a time.

Advantages:
- Fast updates
- Escapes local minima

Disadvantages:
- Noisy gradients



## Mini-Batch Gradient Descent

Uses small subsets of data.

Advantages:
- Efficient
- Stable
- GPU friendly

This dominates modern deep learning.



## Comparison Table

| Method | Speed | Stability | Memory |
|----------|---------|---------|---------|
| Batch GD | Slow | High | High |
| SGD | Fast | Low | Low |
| Mini-Batch | Balanced | High | Moderate |


```

import numpy as np

data=np.arange(100)

batch_size=10

batches=[data[i:i+batch_size] for i in range(0,len(data),batch_size)]

print("Number of batches:",len(batches))
```


# 3. Epochs and Iterations

## Epoch

## Iteration

One parameter update.

Formula:

Iterations = Dataset Size / Batch Size


```

dataset_size=1000
batch_size=100

iterations=dataset_size//batch_size

print("Iterations per Epoch:",iterations)
```

# 5. Why Backpropagation Is Needed

Forward propagation generates predictions.

However training requires gradients.

Brute-force differentiation becomes computationally expensive.

Backpropagation computes gradients efficiently using the chain rule.



# 6. Chain Rule Review

If:

\[
y=f(u)
\]

and

\[
u=g(x)
\]

Then:

\[
\frac{dy}{dx}=\frac{dy}{du}\frac{du}{dx}
\]

This principle powers backpropagation.



# 7. Backpropagation Fundamentals

Historical Contributors:

- David Rumelhart
- Geoffrey Hinton
- Ronald Williams

Backpropagation efficiently computes gradients layer by layer.

Loss:

\[
L
\]

Output:

\[
\hat y
\]

Goal:

\[
\frac{\partial L}{\partial W}
\]

and

\[
\frac{\partial L}{\partial b}
\]

for every parameter.



# 8. Backpropagation Through a Single Neuron

Forward:

\[
z=wx+b
\]

\[
a=\sigma(z)
\]

Loss:

\[
L=(y-a)^2
\]

Backward:

\[
\frac{\partial L}{\partial w}
=
\frac{\partial L}{\partial a}
\frac{\partial a}{\partial z}
\frac{\partial z}{\partial w}
\]

Weight Update:

\[
w=w-\eta\frac{\partial L}{\partial w}
\]



# 9. Multi-Layer Backpropagation

Forward:

\[
Z_1=XW_1+b_1
\]

\[
A_1=f(Z_1)
\]

\[
Z_2=A_1W_2+b_2
\]

Backward:

\[
\delta_2=\frac{\partial L}{\partial Z_2}
\]

\[
\delta_1=\delta_2W_2^T f'(Z_1)
\]

Gradients:

\[
\frac{\partial L}{\partial W_2}=A_1^T\delta_2
\]

\[
\frac{\partial L}{\partial W_1}=X^T\delta_1
\]



# 10. Numerical Example

Input = 1

Weight = 0.5

Bias = 0.1

Forward:

z = 0.6

Sigmoid(z) ≈ 0.645

Loss computed from target.

Gradient calculated using chain rule.

Weight updated to reduce loss.


```

import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

x=1.0
w=0.5
b=0.1

z=w*x+b
a=sigmoid(z)

print("Output:",a)
```


# 11. NumPy Implementation

Forward propagation, loss computation, backpropagation, and parameter updates from scratch.


```

import numpy as np

X=np.array([[1]])
y=np.array([[1]])

W=np.array([[0.5]])
b=np.array([[0.1]])

lr=0.1

def sigmoid(x):
    return 1/(1+np.exp(-x))

for _ in range(5):

    z=X@W+b
    a=sigmoid(z)

    loss=np.mean((y-a)**2)

    dz=(a-y)*a*(1-a)

    dW=X.T@dz
    db=np.sum(dz)

    W-=lr*dW
    b-=lr*db

print("Final Loss:",loss)
```


# 12. Gradient Flow

Gradients move backward through layers.

Strong gradients:
- Fast learning

Weak gradients:
- Slow learning


```

import matplotlib.pyplot as plt
import numpy as np

layers=np.arange(1,11)

strong=np.linspace(1,0.5,10)
weak=np.linspace(1,0.01,10)

plt.figure(figsize=(6,4))
plt.plot(layers,strong,label="Strong Gradient")
plt.plot(layers,weak,label="Weak Gradient")
plt.legend()
plt.grid(True)
plt.title("Gradient Flow")
plt.show()
```


# 13. Vanishing Gradients

Occurs when gradients become extremely small.

Common with:

- Sigmoid
- Tanh

Effects:

- Slow learning
- Early layers stop updating


```

import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-10,10,500)

sigmoid=1/(1+np.exp(-x))
grad=sigmoid*(1-sigmoid)

plt.figure(figsize=(6,4))
plt.plot(x,grad)
plt.title("Sigmoid Gradient")
plt.grid(True)
plt.show()
```
