## Mathematical Foundation

$z = wx + b$  
$a = f(z)$  
Forward propagation: $a=f(WX+b)$

```
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1/(1+np.exp(-x))
```

## Level 1: Single Neuron using NumPy

```
inputs=np.array([2,3])
weights=np.array([0.5,0.8])
bias=0.2
z=np.dot(inputs,weights)+bias
print('Weighted Sum:',z)
print('Output:',sigmoid(z))
```

## Level 2: Manual Forward Propagation

```
X=np.array([1,2])
W=np.array([[0.5,0.8],[0.2,0.3]])
b=np.array([0.1,0.2])
a=sigmoid(np.dot(X,W)+b)
print(a)
```

## Level 3: Multi Layer Perceptron using NumPy

```
np.random.seed(42)
X=np.array([[0,0],[0,1],[1,0],[1,1]])
W1=np.random.randn(2,4)
b1=np.zeros((1,4))
W2=np.random.randn(4,1)
b2=np.zeros((1,1))
hidden=sigmoid(np.dot(X,W1)+b1)
output=sigmoid(np.dot(hidden,W2)+b2)
print(output)
```
